## Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

## Univariate -> "one variable" data visualization

```python
# strip plot
help(sns.stripplot)
```

```python
help(plt.scatter)
```

```python
# strip plot
sns.stripplot(y = 'tip', data=tips, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
# Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
# Swarm plot
help(sns.swarmplot)
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(data=tips, x='day', y='tip', hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
# changing orientation
sns.swarmplot(data=tips, x='tip', y='day', hue='sex', orient='h')
plt.xlabel('tip ($)')
plt.show()
```

## Box and Violin Plot

```python
# plt.subplot(1,2,1)
sns.boxplot(data=tips, x='day', y='tip')
plt.ylabel('tip ($)')
```

```python
# plt.subplot(1,2,2)
sns.violinplot(data=tips, x='day', y='tip')
plt.ylabel('tip ($)');
```

```python
plt.subplot(1,2,1)
sns.boxplot(data=tips, x='day', y='tip')
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(data=tips, x='day', y='tip')
plt.ylabel('tip ($)');
```

```python
# combining plots

sns.violinplot(data=tips, x='day', y='tip', inner=None, color='lightgray')
sns.stripplot(data=tips, x='day', y='tip', size=4, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate -> "Two variables" data visualization

```python
# Join plot

sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show();
```

```python
help(sns.pairplot)
```

```python
# density plot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
plt.show();
```

```python
# pair plot 
# help(sns.pairplot)
sns.pairplot(tips)
plt.show();
```

```python
sns.pairplot(data=tips, hue='sex')
```
